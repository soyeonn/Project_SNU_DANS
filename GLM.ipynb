# Logistic Regression

## Call Data

```python
import pandas as pd

Data = pd.read_csv('/home/soyeon/Project_DANS/Info.csv')
```

```python
Data_A = Data[:21]
Data_B = Data[21:]
```

```python
import numpy as np

Data_A['group'] = np.ones(21)
Data_B['group'] = np.zeros(35)
```

```python
var = [Data_A, Data_B]
Data = pd.concat(var)
Data
```

## information data

```python
info = pd.DataFrame({'ID' : Data['ID'],
                     'Info' : Data['Info'],
                     'group' : Data['group']})
info
```

```python
x = info['Info']
y = info['group']
```

```python
import statsmodels.api as sm

logit_model=sm.Logit(y,x)
result=logit_model.fit()
print(result.summary2())
```

```python
import seaborn as sns
# ci is estimated using bootstrap
sns.regplot(x='Info', y='group', data=info, logistic=True)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.cross_validation import train_test_split

x = info['Info'].values
y = info['group'].values
x = x.reshape(-1, 1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=0)
y_test = y_test.tolist()
y_train = y_train.tolist()
```

```python
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt 

logit_roc_auc = roc_auc_score(y_test, logreg.predict(x_test))
fpr, tpr, thresholds = roc_curve(y_test, logreg.predict_proba(x_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Info')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```
